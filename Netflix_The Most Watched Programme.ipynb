```python
#importing package panda and csv
import pandas as pd
import csv
```

```python
#Assign the csv file to DF and print top 5
df = pd.read_csv('C:/Users/nursyahirah.abrahman/Desktop/netflix-report/CONTENT_INTERACTION/ViewingActivity.csv')
df.head()
```

```python
# LOC functions to search Title col that contain any word 'Under'
df.loc[df['Title'].str.contains("Under",case=False)]
```

```python
#Check for the datatype
df.dtypes
```

```python
#Convert Title column to string datatype
df['Title'] = df['Title'].convert_dtypes()
```

```python
#Check dataype
df.dtypes
```

### SPLIT TITLE INTO 2 COLUMNS, TO EXTRACT ONLY THE TITLE 

```python
#Extract the title up until first ':' found
df1 = df["Title"].str.split(":",n=1,expand=True)
df["Title1"] = df1[0]
df["Title2"] = df1[1]
```

```python
df.head(10)
```

```python
df.loc[df['Title1'].str.contains("Emily",case=False)]
```

```python
#Cheking shape of the data
df.shape
```

### Convert Start Time column to utc datatype

```python
df['Start Time'] = pd.to_datetime(df['Start Time'],utc=True)
df.dtypes
```

```python
#Set the index and convert in to local time where you reside
df = df.set_index('Start Time')
df.index = df.index.tz_convert('Asia/Singapore')
df = df.reset_index()
df.head(1)
```

```python
#Convert Duration col to time_Delta and check the dataype
df['Duration'] = pd.to_timedelta(df['Duration'])
df.dtypes
```

```python
#Choose only important column
```

```python
#Create new column name weekday,hour and minutesl to extract . But the highlight here is col minutes l that convert time in duration columns to seconds
df['weekday'] = df['Start Time'].dt.weekday
df['hour'] = df['Start Time'].dt.hour
df['minutesl'] = df['Duration'].dt.seconds
```

```python
df['Supplemental Video Type'].unique()
```

```python
#Filter Supplemental Video Type to only NaN
to_remove = ['HOOK','RECAP','TEASER_TRAILER','PROMOTIONAL','PREVIEW']
df = df[~df['Supplemental Video Type'].isin(to_remove)]
```

```python
df.head()
```

```python
df.loc[df['Title1'].str.contains("Now You See Me",case=False)]
```

```python
df.head()
```

### By using Title and minutesl columns, groupby and sum the time to see which Movie has highest duration

```python
df2 = df.groupby(['Title1'])['minutesl'].sum().sort_values(ascending=False)
```

```python
#It's a series data
df2
```

```python
#Convert df2 series into dataframe
df3 = pd.DataFrame(df2)
```

```python
df3.head()
```

### Visualize the output above to see in bar plot the top 10 most watched programme

```python
%matplotlib inline
import matplotlib
```

```python
#Take top 10
df4 = df3.head(10)
```

```python
df4.plot(kind='bar', figsize=(20,10),colormap = 'Spectral' ,title='The Most Watched Programme In Netflix', xlabel='The Programme',ylabel = 'Time Spent- in seconds')
```
